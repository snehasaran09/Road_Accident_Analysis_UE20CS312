# Road Accident Analysis per Vehicle Type In India

This notebook consists of a comprehensive analysis of the data collected from the government websit of India.
The dataset consists of Types of vehicles with severity of their accidents belonging to all states and union territories of India.  
In 2021 it was revealed that 2 wheelers accounted for maximum fatal accidents followed by cars. The main aim of this notebook is to detect the vehicles that contribute to fatal accidents in car/jeeps/vans and motorised two wheelers.

As seen in the heatmap , 4 most correlated vehicles types are chosen. This is done to find out how these vehicles cause major accidents to other vehicles. We limit our models to find the severity of motorised two wheeler and cars/jeep/vans only. 
The models are : 
1. Fatal motorvehicle accidents that are caused by tempos/tractors
2. Fatal motorvehicle accidents caused by truck/lorries
3. Fatal car/jeep/van accidents caused by buses
4. Fatal car/jeep/van accidents cause by trucks/lorries

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

1. Fatal motorvehicle accidents caused by tempos/tractors

```python
y = data['Motorised two Wheeler - Number of Road Accidents - Fatal']
x = data.iloc[:,31:36]
```

```python
regressor =  LinearRegression()
xtrain, xtest, ytrain, ytest = train_test_split(x,y,train_size=0.5,test_size=0.5)
```

```python
regressor.fit(xtrain,ytrain)
```

```python
ypred = regressor.predict(xtest)
```

```python
mean_squared_error(ytest,ypred)
```

```python
print('Train Score: ',regressor.score(xtrain,ytrain))
print('Test Score: ',regressor.score(xtest,ytest))
```

```python
r2_score(ytest,ypred)
```
